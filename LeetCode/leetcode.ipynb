```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from pylab import mpl  
mpl.rcParams['font.sans-serif'] = ['FangSong']
mpl.rcParams['axes.unicode_minus'] = False


%matplotlib inline
```

```python
df_problems = pd.read_csv('df_problems.csv')

df_problems['count_pass'] = df_problems['count_pass'].apply(lambda x:x.replace(',','')).astype(int)
df_problems['count_commit'] = df_problems['count_commit'].apply(lambda x:x.replace(',','')).astype(int)
df_problems['ratio_pass'] = df_problems['count_pass'] / df_problems['count_commit']
```

```python
df_problems.head()
```

```python
df_problems.sort_values(by='count_star',ascending=False).head(10)
```

```python
for col in ['count_star','count_pass','count_commit','ratio_pass']:
    print(col)
    plt.figure(figsize=(20,8))
    sns.distplot(df_problems[col])
    sns.distplot(df_problems[df_problems['difficult'] == '简单'][col])
    sns.distplot(df_problems[df_problems['difficult'] == '中等'][col])
    sns.distplot(df_problems[df_problems['difficult'] == '困难'][col])
    plt.legend(['总体','简单','中等','困难'])
    plt.title(col)
    plt.show()
```

```python
sns.barplot(x='difficult',y='count_star',data=df_problems)
plt.tight_layout()  
plt.show()
```

```python
plt.figure(figsize=(16,10))
plt.subplot(3,2,1)
plt.scatter(df_problems['count_star'],df_problems['count_pass'])
plt.title('count_star —— count_pass')
plt.subplot(3,2,2)
plt.scatter(df_problems['count_star'],df_problems['count_commit'])
plt.title('count_star —— count_commit')
plt.subplot(3,2,3)
plt.scatter(df_problems['count_star'],df_problems['ratio_pass'])
plt.title('count_star —— ratio_pass')
plt.subplot(3,2,4)
plt.scatter(df_problems['count_pass'],df_problems['count_commit'])
plt.title('count_pass —— count_commit')
plt.subplot(3,2,5)
plt.scatter(df_problems['count_pass'],df_problems['ratio_pass'])
plt.title('count_pass —— ratio_pass')
plt.subplot(3,2,6)
plt.scatter(df_problems['count_commit'],df_problems['ratio_pass'])
plt.title('count_commit —— ratio_pass')

plt.show()
```
